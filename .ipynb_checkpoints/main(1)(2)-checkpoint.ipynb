# Utils

```python
class TorchDatasetWrapper(torch.utils.data.Dataset):
    def __init__(self, hf_dataset, transform=None):
        self.hf_dataset = hf_dataset
        self.transform=transform

    def __repr__(self):
        return str(self.hf_dataset)

    def __len__(self):
        return len(self.hf_dataset)

    def __getitem__(self, idx):
        example = self.hf_dataset[idx]
        image = example['image']
        if self.transform:
            image =  self.transform(image)
        return image, example['label']
```

```python
def print_patches(image_unfolded):
    import matplotlib.pyplot as plt
    from torchvision.utils import make_grid
    if len(image_unfolded.shape) == 5:
        image_unfolded = image_unfolded[None,:,:,:,:,:]
    for i in range(image_unfolded.shape[0]):
        n_channels, n_patches_h, n_patches_w, h, w = image_unfolded.shape[1:]
        patches = image_unfolded[i].permute(1, 2, 0, 3, 4).reshape(-1, n_channels, h, w)
        grid = make_grid(patches, nrow=n_patches_w, padding=2, normalize=False)
        plt.figure(figsize=(3, 3))
        plt.imshow(grid.permute(1, 2, 0).numpy())
        plt.axis('off')
        plt.show()
```

```python
def gen_super_tiny(dataset, split, q=10, p=0, step=500, classes=200):
    from datasets import Dataset
    images_per_class = {"validation":50,
                       "test":50,
                       "train":500
                       }
    dataset_length = images_per_class[split]*classes
    t = [dataset[p+i:i+q] for i in range(0, dataset_length-q, step)]
    all_images = [image for image_class_dict in t for image in image_class_dict["image"]]
    all_labels = [image for image_class_dict in t for image in image_class_dict["label"]]
    return Dataset.from_dict({"image":all_images, "label":all_labels})
```

# Network

#### Make some tests with the super tiny dataset

```python
class ViT(torch.nn.Module):
    def __init__(self, num_classes, in_channels, image_size, patch_size, hidden_size, layers, attention_heads):
        super().__init__()
        self.num_classes = num_classes  
        self.in_channels = in_channels  
        self.image_size = image_size  
        self.patch_size = patch_size  
        self.patch_area = patch_size*patch_size*in_channels
        self.num_patches = int((self.image_size/self.patch_size)**2)
        self.hidden_size = hidden_size  
        self.layers = layers
        self.attention_heads = attention_heads

        init_data = f"""
        ViT Initialized with parameters:
- num_classes: {self.num_classes}
- in_channels: {self.in_channels}
- image_size: {self.image_size}
- patch_size: {self.patch_size}
- num_patches: {self.num_patches}
- patch_area: {self.patch_area}
- hidden_size: {self.hidden_size}
- layers: {self.layers}
- attention_heads: {self.attention_heads}
        """
        print(init_data.strip())
        
        #self.embeddings = torch.nn.Conv2d(in_channels, out_channels = 1, kernel_size = patch_size, stride = patch_size)
        #self.patch_embedding_size = (int((image_size * (patch_size - 1) - 1) / patch_size)+1)**2
        self.linear_patch_embeddings = torch.nn.Conv2d(in_channels=in_channels, out_channels=hidden_size, kernel_size=patch_size, stride=patch_size)
        self.cls_token = torch.nn.Parameter(torch.randn(1, 1, self.hidden_size))
        self.pos_embedding = torch.nn.Parameter(torch.randn(1, self.num_patches + 1, self.hidden_size))
        self.encoder_block = torch.nn.TransformerEncoderLayer(self.hidden_size, attention_heads, dim_feedforward=4*self.hidden_size, batch_first=True)
        self.encoder = torch.nn.TransformerEncoder(self.encoder_block, layers, norm=torch.nn.LayerNorm(self.hidden_size))
        self.head = torch.nn.Linear(self.hidden_size, self.num_classes)
    
    def forward(self, x):
        #x = x.unfold(2, self.patch_size, self.patch_size).unfold(3,self.patch_size,self.patch_size) #(batch_size, channels, patch, patch, patch_pixels, patch_pixels)
        #x = x.permute(0, 2, 3, 1, 4, 5) #(batch_size, patch, patch, channels, patch_pixels, patch_pixels)
        #x = x.flatten(start_dim=3) #(batch_size, patch, patch, patch_area * channels
        #x = x.flatten(1, 2) #(batch_size, number_patches, patch_area * channels
        x = self.linear_patch_embeddings(x) #(batch_size, hidden_size, number_patches, number_patches
        x = x.flatten(start_dim=2).permute(0,2,1)
        cls_tokens = self.cls_token.expand(x.shape[0], -1, -1) #(batch_size, number_patches^2 + 1, hidden_size
        x = torch.cat((cls_tokens, x), dim=1)  # (batch_size, number_patches^2 + 1, hidden_size)
        x = x + self.pos_embedding
        x = self.encoder(x)  # (batch_size, number_patches^2 + 1, hidden_size)
        cls_output = x[:, 0, :] 
        return self.head(cls_output)
```

```python
parameters = {
    'num_classes': 200,
    'in_channels': 3,
    'image_size': 64,
    'patch_size': 16,
    'hidden_size': 768,
    'layers': 12, 
    'attention_heads': 12 
}
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = ViT(**parameters)
model.to(device)
model.eval()
with torch.no_grad():
    attn_output_weights = model(torch.randn(2, 3, 64, 64).to(device))
    print(attn_output_weights)
```

# Train

```python
from torchmetrics import Accuracy, F1Score, Precision, Recall, ConfusionMatrix
import torch
from tqdm import tqdm

def run_epoch(epoch, num_epochs, device, optimizer, loss_fn, model, data_loader, num_classes, train, confusion_matrix=False):
    name = "train" if train else "validation"
    model.train() if train else model.eval()
    running_loss = 0.0

    # Initialize metrics
    accuracy_metric = Accuracy(task="multiclass", num_classes=num_classes).to(device)
    if confusion_matrix:
        cm_metric = ConfusionMatrix(task="multiclass", num_classes=num_classes).to(device)

    for i, data in enumerate(data_loader):
        images, labels = data
        images, labels = images.to(device), labels.to(device)
        
        with torch.set_grad_enabled(train):
            preds_logits = model(images)
            loss = loss_fn(preds_logits, labels)
        
        running_loss += loss.item()
        if train:
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        predicted_labels = torch.argmax(preds_logits, dim=1)
        accuracy_metric.update(predicted_labels, labels)
        if confusion_matrix:
            cm_metric.update(predicted_labels, labels)

        print(f"Epoch {epoch} Batch {i+1} loss: {running_loss/(i+1)} accuracy: {accuracy_metric.compute().item()}")

    avg_loss = running_loss / len(data_loader)
    accuracy = accuracy_metric.compute().item()
        
    return avg_loss, accuracy, 0, 0, 0
```

```python
from tqdm import tqdm
import torch

def train_model(model, train_loader, val_loader, num_epochs, learning_rate, model_save_path, num_classes):
    """
    Train a PyTorch model with the given parameters.
    
    Args:
        model: The PyTorch model to train
        train_loader: DataLoader for training data
        val_loader: DataLoader for validation data
        num_epochs: number of training epochs
        learning_rate: learning rate for optimizer
        model_save_path: path to save trained model
        num_classes: number of classes for classification

    Returns:
        Tuple of training and validation metrics including losses, F1 scores, precisions, recalls, and accuracies.
    """
    patience = 10
    toggle = 0
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    loss_fn = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    best_vloss = float('inf')
    
    train_losses, train_f1s, train_precisions, train_recalls, train_accs = [], [], [], [], []
    val_losses, val_f1s, val_precisions, val_recalls, val_accs = [], [], [], [], []
    
    model = model.to(device)
    
    # Training loop
    for epoch in range(num_epochs):
        # Training phase
        train_avg_loss, train_acc, train_f1, train_precision, train_recall = run_epoch(
            epoch, num_epochs, device, optimizer, loss_fn, model, train_loader, num_classes, train=True, confusion_matrix=True
        )
        train_losses.append(train_avg_loss)
#        train_f1s.append(train_f1)
#        train_precisions.append(train_precision)
#        train_recalls.append(train_recall)
        train_accs.append(train_acc)
        
        val_avg_loss, val_acc, val_f1, val_precision, val_recall = run_epoch(
            epoch, num_epochs, device, optimizer, loss_fn, model, val_loader, num_classes, train=False, confusion_matrix=True
        )
        val_losses.append(val_avg_loss)
#        val_f1s.append(val_f1)
#        val_precisions.append(val_precision)
#        val_recalls.append(val_recall)
        val_accs.append(val_acc)
        
        print(f"Epoch {epoch+1}/{num_epochs}")
        print(f"Train Loss: {train_avg_loss:.4f} | Val Loss: {val_avg_loss:.4f}")
        print(f"Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
        print("-----------------------------------")
        
        # Save model if validation loss improves
        if val_avg_loss < best_vloss:
            toggle = 0 if toggle else 0
            not_increase = 0
            best_vloss = val_avg_loss
            if not os.path.isdir(MODEL_DIRS): os.mkdir("data/models")
            torch.save(model.state_dict(), model_save_path[:-4]+str(toggle)+".pth")

        if not_increase >= patience:
            print(f"Enough! Best model saved in {model_save_path[:-4]+str(toggle)+'.pth'}")
            break
    
    return (
        train_losses, train_f1s, train_precisions, train_recalls, train_accs,
        val_losses, val_f1s, val_precisions, val_recalls, val_accs
    )

# Note: Ensure the 'run_epoch' function is defined and returns the necessary metrics.
```
